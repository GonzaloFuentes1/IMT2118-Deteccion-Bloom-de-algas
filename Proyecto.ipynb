# Timelapse

Para una visualización más interesante del indice de OC4ME, se utilizaron las imagenes descargadas en formato .tif y se creo un timelapse con la libreria PIL, junto a cargar las imagenes con rasterio. 

##### Las visualizaciones atractivias estan guardadas como "timelapse_quellon.gif" y "timelapse_villarica.gif" en la carpeta.

```python
from matplotlib.colors import LinearSegmentedColormap
import rasterio
import matplotlib.pyplot as plt
import imageio
import os
from PIL import Image

# Lista de nombres de archivo de las imágenes .tif
file_names = ["Quellón/MARZO-02-2020.tif",
              "Quellón/MARZO-16-2020.tif","Quellón/MARZO-21-2020.tif","Quellón/MARZO-26-2020.tif","Quellón/MARZO-31-2020.tif",
              "Quellón/ABRIL-08-2020.tif"]
sns.set_style("white")
sns.despine(left=True, bottom=True)

colors = [
    (0.0, '#FF0000'),
    (0.000001, '#FFA07A'),
    (0.0005, '#FFC0CB'),
    (0.001, '#FF6699'), 
    (0.005, '#FFFF00'),  
    (0.01, '#00FF00'),
    (0.02, '#99CCFF'), 
    (0.05, '#77AADD'), 
    (0.1, '#6699FF'), 
    (0.15, '#5588EE'), 
    (0.2, '#3366FF'),
    (0.4, '#1155DD'),
    (0.6, '#0033CC'),
    (0.8, '#002299'),
    (1, '#0000FF')   
]

cmap = LinearSegmentedColormap.from_list('chla', colors)
# Crear una lista para almacenar los cuadros del timelapse
frames = []

# Iterar sobre cada imagen y agregarla al timelapse
for i, file_name in enumerate(file_names):
    # Abrir la imagen con Rasterio
    src = rasterio.open(file_name)

    # Leer los datos de la banda
    data = src.read(1)

    # Configurar el tamaño de la figura sin marco
    fig = plt.figure(frameon=False)

    # Agregar la imagen al timelapse sin un fondo definido
    plt.imshow(data, cmap=cmap)

    # Configurar título y etiqueta del eje
    plt.title("Quellón: Marzo-Febrero de 2020",color = "black")
    plt.xlabel("Frame {}".format(i + 1))
    # Eliminar los ejes y márgenes
    plt.axis('off')
    plt.margins(0, 0)


    # Guardar la imagen actual en un archivo temporal
    temp_file_path = "frame{}.png".format(i + 1)
    plt.savefig(temp_file_path, transparent=True,dpi =300)
    plt.close(fig)

    # Abrir la imagen temporal y hacer el fondo invisible
    img = Image.open(temp_file_path).convert("RGBA")
    data = img.getdata()

    new_data = []
    for item in data:
        if item[:3] == (255, 255, 255):  # Verificar píxeles blancos
            new_data.append((255, 255, 255, 0))  # Establecer fondo transparente
        else:
            new_data.append(item)

    img.putdata(new_data)

    # Guardar la imagen con fondo invisible en la lista de cuadros
    frames.append(img)

    # Eliminar el archivo temporal
    os.remove(temp_file_path)

# Replicar los cuadros para hacer que el GIF esté en bucle
frames.extend(frames[::-1])

# Guardar el timelapse como un archivo GIF en bucle con una duración de 2 segundos por imagen
imageio.mimsave("timelapse_quellon.gif", frames, duration = 2000, loop=10)

# Eliminar los archivos temporales de los cuadros
for i, _ in enumerate(file_names):
    file_path = "frame{}.png".format(i + 1)
```

# Análisis de regresión


Se leen los imagenes de OC4ME, precipitaciones y temperaturas aproximadas para el 31 de marzo de 2020, guardando en una grilla 1000x1000 los datos a utilizar en al regresión.

```python
import pandas as pd
import rasterio

# Ruta de la imagen .tif
ruta_imagen = "MARZO-31-2020.tif"

# Abrir la imagen con rasterio
with rasterio.open(ruta_imagen) as dataset:
    # Leer la banda de la imagen
    banda = dataset.read(1)

    # Obtener el tamaño de la imagen
    filas = dataset.height
    columnas = dataset.width

    # Crear una lista de coordenadas
    coordenadas = []
    for i in range(1000):
        for j in range(1000):
            coordenadas.append((i, j))

    # Crear una lista de valores de los vértices
    valores_vertices = []
    for coord in coordenadas:
        i, j = coord
        valores_vertices.append(banda[i * filas // 1000][j * columnas // 1000])

    # Crear el DataFrame
    df1 = pd.DataFrame({'Coordenadas': coordenadas, 'Chla': valores_vertices})

# Ruta de la imagen .tif
ruta_imagen = "PRECIPITACIONES-31-03-2020.tif"

# Abrir la imagen con rasterio
with rasterio.open(ruta_imagen) as dataset:
    # Leer la banda de la imagen
    banda = dataset.read(1)

    # Obtener el tamaño de la imagen
    filas = dataset.height
    columnas = dataset.width

    # Crear una lista de coordenadas
    coordenadas = []
    for i in range(1000):
        for j in range(1000):
            coordenadas.append((i, j))

    # Crear una lista de valores de los vértices
    valores_vertices = []
    for coord in coordenadas:
        i, j = coord
        valores_vertices.append(banda[i * filas // 1000][j * columnas // 1000])

    # Crear el DataFrame
    df2 = pd.DataFrame({'Coordenadas': coordenadas, 'Precipitaciones_mm': valores_vertices})

    # Ruta de la imagen .tif
ruta_imagen = "TEMPERATURA-31-03-2020.tif"

# Abrir la imagen con rasterio
with rasterio.open(ruta_imagen) as dataset:
    # Leer la banda de la imagen 
    banda = dataset.read(1)

    # Obtener el tamaño de la imagen
    filas = dataset.height
    columnas = dataset.width

    # Crear una lista de coordenadas
    coordenadas = []
    for i in range(1000):
        for j in range(1000):
            coordenadas.append((i, j))

    # Crear una lista de valores de los vértices
    valores_vertices = []
    for coord in coordenadas:
        i, j = coord
        valores_vertices.append(banda[i * filas // 1000][j * columnas // 1000])

    # Crear el DataFrame
    df3 = pd.DataFrame({'Coordenadas': coordenadas, "Temperatura": valores_vertices})
```

Unimos los 3 dataframes de información de las imagenes y filtramos los datos nulos (como la imagen tiene partes faltantes que corresponden a tierra, estos valores podria afectar a la predicción). También se calcula la variable de clorofila, pero en su logaritmo, puesot que podria mejorar la regresión considerando la naturaleza de la variable del algoritmo de OC4ME.

```python
import numpy as np
mergeado = df1.merge(df2).merge(df3)
mergeado.fillna(0, inplace=True)
filtro1 = mergeado[mergeado["Chla"]>0]
filtro2 = filtro1[filtro1["Precipitaciones_mm"]>0]
filtro3 = filtro2[filtro2["Temperatura"]>0]
filtro3['log_chla'] = np.log10(filtro3["Chla"])
```

Para probar, si crea un modelo de regresión simple sin dependencia espacial de ningun tipo. 

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# Supongamos que tienes un DataFrame llamado df con columnas 'Variable_independiente' y 'Variable_dependiente'
X = filtro3[["Precipitaciones_mm","Temperatura"]]
y = filtro3['Chla']

# Crear el modelo de regresión lineal
modelo = LinearRegression()

# Ajustar el modelo a los datos
modelo.fit(X, y)

# Imprimir los coeficientes y la intercepción
print('Coeficientes:', modelo.coef_)
print('Intercepción:', modelo.intercept_)

# Después de ajustar el modelo
r_squared = modelo.score(X, y)
print("Coeficiente de determinación R^2:", r_squared)
```
